```python
data = {"group": [0 ,0 ,0 ,1 ,1 ,1 ,2 ,2 ,2 ,3 ,3 ], 
            "batch": [0 ,1 ,2 ,0 ,1 ,2 ,0 ,1 ,2 ,0 ,1 ], 
            "count": [10,12,8 ,20,5 ,24,12,46,3 ,17,5 ]}
res = []
for k, lst in data.items():
    temp = []
    for v in lst:
        temp.append({k: v})
    res.append(temp)
```

```python
new_data = res[0]
for i in [1,2]:
    for j in range(len(new_data)):
        new_data[j].update(res[i][j])
```
